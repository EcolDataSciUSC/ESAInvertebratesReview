```python
# Load Python packages
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
# Read in the .csv file with the subset of documentation for testing
my_docs = pd.read_csv("../data/pdf_all_August2024.csv")
```

```python
import time

# Save the PDF files for further processing by ChatGPT
i = 0
for url in my_docs["urls"]:

    response = requests.get(url)
    print(response.status_code)

    if response.status_code == 200:
        with open(f"../data/samplePDFs/{i}.pdf", "wb") as file:
            file.write(response.content)
    
    i = i+1
    time.sleep(2)
```
